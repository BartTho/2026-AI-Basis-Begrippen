## Sigmoid

output is altijd een getal tussen 0 en 1. Deze gebruiken we als output voor klastificatie

```python
def sigmoid(x):
    return 1 / (math.exp(-x))
```

```python
# Genereer x-waarden van -10 tot 10
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))

# Plot de grafiek
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
print(sigmoid(-100), sigmoid(-100),sigmoid(-10), sigmoid(0),  sigmoid(1), sigmoid(-1), sigmoid(10), sigmoid(100))
```

## tanh

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))  
```

```python
# Genereer x-waarden van -10 tot 10
x = np.linspace(-10, 10, 100)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Plot de grafiek
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
print(tanh(-100),tanh(-10),tanh(-1), tanh(0),  tanh(1),tanh(10),tanh(100))
```

## ReLu

```python
def ReLu(x):
    return max(0, x)
```

```python
# Genereer x-waarden van -10 tot 10
x = np.linspace(-10, 10, 100)
y = np.maximum(0, x)

# Plot de grafiek
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
print(ReLu(-100), ReLu(-10),ReLu(-1),ReLu(0),  ReLu(1), ReLu(10), ReLu(100))
```

```python
def leaky_ReLu(x):
    return max(0.01 * x, x)
```

```python
# Genereer x-waarden van -10 tot 10
x = np.linspace(-20, 10, 100)
y = np.maximum(0.01 * x, x)

# Plot de grafiek
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
print( leaky_ReLu(-100), leaky_ReLu(10), leaky_ReLu(0),  leaky_ReLu(1), leaky_ReLu(10), leaky_ReLu(100))
```
